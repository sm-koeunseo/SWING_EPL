```python
import pandas as pd
```

```python
df = pd.read_csv('CLM_SPI_DD_20230712115724.csv')
```

```python
df['일시'].head()
```

```python
df = df.copy()
df.loc[:, '일시'] = pd.to_datetime(df['일시'], format='%Y.%m.%d').dt.strftime('%Y-%m-%d')
df['일시'] = pd.to_datetime(df['일시'])
```

```python
df['일시'].head()
```

```python
df['년월'] = df['일시'].dt.strftime('%Y-%m')
df['년월'].head()
```

```python
data = df.groupby(['지점','지점명','년월'])[['SPI3','SPI6']].mean().reset_index()
```

```python
data
```

```python
data.to_csv('SPI.csv', index=False, encoding='utf-8')
```

```python
dt = data[data['지점명']=='광주']
dt
```

```python
import matplotlib.pyplot as plt

x = dt['년월']
y1 = dt['SPI3']
y2 = dt['SPI6']
```

```python
plt.plot(x, y1, marker='o', label='SPI3')
plt.plot(x, y2, marker='o', label='SPI6')
plt.xlabel('month')
plt.ylabel('value')
plt.legend()
plt.show()
```
