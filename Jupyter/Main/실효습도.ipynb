```python
import pandas as pd
df_eh = pd.read_csv('실효습도.csv', encoding = 'cp949')
```

```python
df = pd.read_csv('산불발생횟수.csv', encoding = 'cp949')
wf13 = df[(df['연도'] == 2013)].shape[0]
wf14 = df[(df['연도'] == 2014)].shape[0]
wf15 = df[(df['연도'] == 2015)].shape[0]
wf16 = df[(df['연도'] == 2016)].shape[0]
wf17 = df[(df['연도'] == 2017)].shape[0]
wf18 = df[(df['연도'] == 2018)].shape[0]
wf19 = df[(df['연도'] == 2019)].shape[0]
wf20 = df[(df['연도'] == 2020)].shape[0]
wf21 = df[(df['연도'] == 2021)].shape[0]
wf22 = df[(df['연도'] == 2022)].shape[0]
```

```python
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 40
plt.rcParams["figure.figsize"] = (60, 10)
```

```python
labels = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
data_wf = [wf13/100, wf14/100, wf15/100, wf16/100, wf17/100, wf18/100, wf19/100, wf20/100, wf21/100]
data_he = [11, 13, 16, 12, 21, 20, 18, 10, 8]

plt.plot(labels, data_wf, marker='o')
plt.plot(labels, data_he, marker='o')
plt.title('Number of Wildfires according to Effective Humidity')
plt.xlabel('Year')
plt.ylabel('Number')
plt.show()

from scipy.stats import pearsonr
correlation_coefficient, p_value = pearsonr(data_wf, data_he)

print("Correlation coefficient:", correlation_coefficient)
print("p-value:", p_value)
```
