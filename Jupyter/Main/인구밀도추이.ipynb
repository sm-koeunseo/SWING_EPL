```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 15
plt.rcParams["figure.figsize"] = (15, 7)

data = {
    '행정구역별': ['경기도', '강원도', '충청남도', '전라남도', '경상북도', '경상남도'],
    2017: [1261.6, 90.4, 262.9, 145.3, 140.7, 317.4],
    2018: [1286.2, 90.3, 265.1, 145.0, 140.4, 317.9],
    2019: [1305.0, 90.3, 265.4, 144.8, 140.2, 317.6],
    2020: [1325.3, 90.4, 264.0, 144.9, 138.9, 316.2],
    2021: [1338.9, 90.4, 263.8, 143.9, 138.4, 313.6]
}

df = pd.DataFrame(data)
df.set_index('행정구역별', inplace=True)

df_transposed = df.T

df_transposed.plot(kind='line')

plt.title('지역별 인구 밀도 추이')
plt.xlabel('연도')
plt.ylabel('인구 밀도')

plt.show()
```
