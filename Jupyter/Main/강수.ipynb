```python
import pandas as pd
df = pd.read_csv('rain.csv', encoding='cp949')
df.head()
```

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 15
plt.rcParams["figure.figsize"] = (30, 8)
```

```python
plt.title("최다월강수량")
plt.xlabel("지역명")
plt.ylabel("월강수량")
plt.bar(df['최다강수량지점'], df['최다월강수량(mm)'], width = 0.5)
```

```python
plt.title("1시간최다강수량")
plt.xlabel("지역명")
plt.ylabel("1시간강수량")
plt.bar(df['1시간최다강수량지점'], df['1시간최다강수량(mm)'], width = 0.5)
```
