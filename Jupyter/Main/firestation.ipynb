```python
data = {
    '시도': ['gyeonggi', 'gangwon', 'chungbuk', 'chungnam', 'jeonbuk', 'jeonnam', 'gyeongbuk', 'gyeongnam'],
    '도로보급률': [1.3, 1.9, 2, 1.8, 2.3, 2.3, 2, 2.2]   
}

# 그래프 생성
plt.figure(figsize=(15, 8))
colors = ['orange', 'purple', 'brown', 'pink', 'olive', 'teal', 'gold', 'navy', 'lime']
plt.bar(data['시도'], data['도로보급률'], color=colors)
plt.xlabel('District', fontsize=20)
plt.ylabel('Road Supply', fontsize=20)
plt.title('Road Supply Rate / Population and Area', fontsize=26)

plt.xticks(fontsize=15)

# y축 범위 설정
plt.ylim(0, 2.5)

# y축 눈금 간격 설정
plt.yticks([i/10 for i in range(0, 26, 3)])  # 0부터 2.5까지 0.3 간격으로 설정

# 그래프 출력
plt.show()
```
